```python
%matplotlib inline  
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set_color_codes("bright")
import glob
from pylab import rcParams

pd.set_option('max_columns', 160)
pd.set_option('max_rows', 100)
```

```python
sns.set(font_scale=1.25)
sns.set_style("ticks")
f, axarr = plt.subplots(2, figsize=(5,8), sharex=True)

gRNAs = pd.read_csv('CDSrc_guides.cleaned.csv',sep=',',header=None,names=['PEG','strand','start_bp','length_bp','frac','seq'])

fracORF=list(range(100))
fracORF = [x/100 for x in fracORF]
fracCoverage = []
for thresh in fracORF:
    fracCoverage.append(gRNAs[gRNAs["frac"]<=thresh].shape[0]/gRNAs.shape[0])


sns.distplot(gRNAs["frac"],kde=False, hist=True, color='black', bins=100, ax=axarr[0])
axarr[0].set(xlabel='', ylabel='Number of gRNAs')
axarr[1].plot(fracORF, fracCoverage, linewidth=1.0, linestyle="--", color="blue")

plt.xlim([-0.01,1])
plt.xlabel('Fraction of ORF')
plt.ylabel('Fraction of all gRNAs')
```
